```python
import requests
import os

class spider:
    def __init__(self,keyword,pages):
        self.keyword=keyword
        self.pages=pages
        self.params=[]
        for i in range(30,30*self.pages+30,30):
            self.params.append({
                'tn':'resultjson_com',
                'ipn':'rj',
                'ct':201326592,
                'is':'',
                'fp':'result',
                'queryWord':self.keyword,
                'cl':2,
                'lm':2,
                'ie':'utf-8',
                'oe':'utf-8',
                'adpicid':'',
                'st':-1,
                'z':'',
                'ic':0,
                'word':self.keyword,
                's':'',
                'se':'',
                'tab':'',
                'width':'',
                'height':'',
                'face':0,
                'istype':2,
                'qc':'',
                'nc':1,
                'fr':'',
                'pn':i,
                'rn':30,
                'gsm':'le',
                '1488942260214':''
            })
            
    def getManyPages(self):
        url='https://image.baidu.com/search/acjson'
        urls=[]
        for i in self.params:
            urls.append(requests.get(url,params=i).json().get('data'))
        return urls
    
    def getImg(self,dataList,localPath):
        if not os.path.exists(localPath):#新建文件夹
            os.mkdir(localPath)
        x=0
        for list in dataList:
            for i in list:
                if i.get('thumbURL') != None:
                    downLoad_url=i.get('thumbURL')
                    print('正在下载：%s' % downLoad_url)
                    ir=requests.get(i.get('thumbURL'))
                    open(localPath + '%d.jpg' % x,'wb').write(ir.content)
                    x+=1
                else:
                    #print('404，图片链接无法访问')
                    break
if __name__=='__main__':
    keyword=input('输入关键词：')
    a=input('输入要下载的页数：')
    n=int(a)
    scrapy=spider(keyword,n)#参数1：关键字，参数2：要下载的页数
    dataList=scrapy.getManyPages()
    scrapy.getImg(dataList,'C:/Users/ASUS/Desktop/test/')#指定保存的路径
```

```python
#打印下载结果
import matplotlib.pyplot as plt # plt 用于显示图片
import matplotlib.image as mpimg # mpimg 用于读取图片
import numpy as np
from scipy import misc #对图像进行缩放

path = 'C:/Users/ASUS/Desktop/test/'
for lists in os.listdir(path):
    sub_path = os.path.join(path, lists)
    print(sub_path)
    lena = mpimg.imread(sub_path) # 读取
    lena_new_sz = misc.imresize(lena, 0.8) # 第二个参数如果是整数，则为百分比，如果是tuple，则为输出图像的尺寸
    plt.imshow(lena_new_sz)
    plt.axis('off') # 不显示坐标轴
    plt.show()
```
